```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0"
}

classes = [
    "امراض-الغدد-الصماء",
    "مرض-السكري",
    "امراض-الجهاز-التنفسي",
    "امراض-الدم",
    "امراض-الاطفال"
]

data = []

for category in classes:

    print("\n" + "=" * 100)
    print(f"Category: {category}")
    print("=" * 100)

    for page in range(1, 200):

        url = f"https://altibbi.com/اسئلة-طبية/{category}?page={page}"

        while True:

            response = requests.get(url, headers=headers)

            if response.status_code == 429:
                print("429 Too Many Requests")
                print("Waiting 30 seconds...\n")
                time.sleep(30)
                continue

            if response.status_code != 200:
                print(f"Error: {response.status_code}")
                break

            break

        soup = BeautifulSoup(response.text, "html.parser")

        questions = soup.find_all(
            "div",
            itemtype="https://schema.org/Question"
        )

        if not questions:
            print(f"No questions found in Page {page}")
            continue

        print(f"\n========== Page {page} ==========\n")

        for i, q in enumerate(questions, 1):

            question = q.find("h2")
            question_text = question.get_text(strip=True) if question else "No Question"

            answer = q.find("div", itemprop="acceptedAnswer")
            answer_text = answer.get_text(" ", strip=True) if answer else "No Answer"


            data.append({
                "Category": category,
                "Page": page,
                "Question_ID": i,
                "Question": question_text,
                "Answer": answer_text
            })


            print(f"Question {i}:")
            print(question_text)
            print("\nAnswer:")
            print(answer_text)
            print("-" * 80)

        time.sleep(2)


df = pd.DataFrame(data)

df.drop_duplicates(
    subset=["Question"],
    inplace=True
)

df.reset_index(drop=True, inplace=True)


print(df.head())

print("Dataset Shape:", df.shape)


df.to_csv(
    "medical_dataset.csv",
    index=False,
    encoding="utf-8-sig"
)

print(f"Total Questions Collected = {len(df)}")
```

```python
df = pd.read_csv("medical_dataset.csv")
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "Qwen/Qwen2.5-1.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
```

```python
def generate_qwen_answer(question):

    messages = [
        {
            "role": "system",
            "content": (
                "أنت مساعد للإجابة عن الأسئلة الطبية. "
                "أجب عن السؤال باللغة العربية بناءً على المعلومات الطبية المتاحة. "
                "قدم إجابة واضحة ومباشرة."
            )
        },
        {
            "role": "user",
            "content": question
        }
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = tokenizer(
        text,
        return_tensors="pt"
    ).to(model.device)

    with torch.no_grad():

        outputs = model.generate(
            **inputs,
            max_new_tokens=200,
            do_sample=False
        )

    generated_tokens = outputs[0][
        inputs["input_ids"].shape[1]:
    ]

    answer = tokenizer.decode(
        generated_tokens,
        skip_special_tokens=True
    )

    return answer.strip()
```

```python
from sentence_transformers import SentenceTransformer

similarity_model = SentenceTransformer(
    "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)
```

```python
def calculate_similarity(
    ground_truth,
    model_answer
):

    embeddings = similarity_model.encode(
        [ground_truth, model_answer],
        convert_to_tensor=True
    )

    score = torch.nn.functional.cosine_similarity(
        embeddings[0].unsqueeze(0),
        embeddings[1].unsqueeze(0)
    ).item()

    return score
```

```python
question = df.iloc[0]["Question"]

qwen_answer = generate_qwen_answer(question)

ground_truth = df.iloc[0]["Answer"]

similarity_score = calculate_similarity(
    ground_truth,
    qwen_answer
)

print("Question:")
print(question)

print("\nQwen Answer:")
print(qwen_answer)

print("\nGround Truth:")
print(ground_truth)

print("\nSimilarity Score:")
print(f"{similarity_score:.4f}")

print("\nSimilarity Percentage:")
print(f"{similarity_score * 100:.2f}%")
```

```python
results = []

for index, row in df.head(10).iterrows():

    question = row["Question"]
    ground_truth = row["Answer"]

    # Generate Qwen answer
    qwen_answer = generate_qwen_answer(question)

    # Calculate similarity
    similarity_score = calculate_similarity(
        ground_truth,
        qwen_answer
    )

    results.append({
        "Question": question,
        "Ground_Truth": ground_truth,
        "Qwen_Answer": qwen_answer,
        "Similarity_Score": similarity_score,
        "Similarity_Percentage": similarity_score * 100
    })

    print(f"Processed {index + 1}/10")
```

```python
results_df = pd.DataFrame(results)

results_df.to_csv(
    "qwen_evaluation_results.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\nAverage Similarity Score:")
print(results_df["Similarity_Score"].mean())

print("\nAverage Similarity Percentage:")
print(results_df["Similarity_Percentage"].mean())
```
